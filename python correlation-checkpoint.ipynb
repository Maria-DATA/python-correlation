```python
iris_datasets.columns
```

```python
iris_datasets=iris_datasets[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
iris_datasets.head()
```

```python
iris_datasets.corr().style.background_gradient(cmap='coolwarm')
```

```python
plt.figure(figsize=(8,8))
seaborn.heatmap(iris_datasets.corr(),annot=True, cmap='coolwarm')
```

```python
iris_datasets_pairplot=iris_datasets.dropna()
iris_datasets.to_excel
seaborn.pairplot(iris_datasets_pairplot,height=1.5)
```

```python
import matplotlib.pyplot as plt 
plt.imshow(iris_datasets.corr(), cmap='coolwarm',interpolation ='none')
plt.colorbar()
plt.xticks(range(len(iris_datasets.columns)), iris_datasets.columns, rotation=90)
plt.yticks(range(len(iris_datasets.columns)), iris_datasets.columns)
plt.gcf().set_size_inches(8,8)

labels=iris_datasets.corr().values
for y in range(labels.shape[0]):
    for x in range(labels.shape[1]):
        plt.text(x,y,'{:.2f})'.format(labels[y,x]),ha='center',va='center',color='white')
 
```
